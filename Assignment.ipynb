## Assignment WebScraping


```python
from bs4 import BeautifulSoup
import requests
import itertools
import pandas as pd
```

```python
url = 'https://www.programmableweb.com/apis/directory'
```

```python

api_dict = {}
api_no = 0
page_no = 0
```

```python

while True:
    response = requests.get(url)
    data = response.text
    soup = BeautifulSoup(data, 'html.parser')
    api_name = soup.find_all("td",{"class":"views-field views-field-title col-md-3"})
    api_category = soup.find_all("td", {"class":"views-field views-field-field-article-primary-category"})
    api_description = soup.find_all("td",{"class":"views-field views-field-search-api-excerpt views-field-field-api-description hidden-xs visible-md visible-sm col-md-8"})
    for name, category, description in zip(api_name, api_category, api_description):
        api = name.find("a").text
        api_url = url + name.find("a").get("href")
        category_name_t = category.find("a")
        category_name = category_name_t.text if category_name_t else 'N/A'
        description_i = description.text
        #print("API : ", api,"\nLink : ", api_url,"\nCategory : ", category_name, "\nDescription : ",description_i)
        api_no += 1
        api_dict[api_no] = [api, api_url, category_name, description_i]
        
    url_tag = soup.find("a",{"title":"Go to next page"})
    if url_tag.get('href') :
        url = 'https://www.programmableweb.com' + url_tag.get('href')
        print(url)
        page_no += 1
        print(page_no)
    else:
        break
```

```python
api_df = pd.DataFrame.from_dict(api_dict, orient = 'index', columns = ['API Name','API URL','API Category', 'API Description'])
```
